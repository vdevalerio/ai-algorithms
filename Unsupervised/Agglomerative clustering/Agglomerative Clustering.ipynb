# Agglomerative Clustering
## O Agglomerative Clustering é um algoritmo de clustering hierárquico que constrói uma hierarquia de clusters. Ele começa com cada ponto como um cluster individual e, iterativamente, agrupa os clusters mais próximos até que todos os pontos pertençam a um único cluster.

### Como funciona:

#### Inicialização:
* Cada ponto é considerado um cluster individual.

#### Agrupamento Iterativo:
* Identificação dos dois clusters mais próximos.
* Fusão desses clusters em um novo cluster.
* É repetido até que todos os pontos estejam em um único cluster.

![agclus](agclus.webp)

#### Vantagens

* Produz uma hierarquia de clusters.
* Não requer a especificação do número de clusters a priori.
* Pode ser visualizado facilmente usando dendrogramas.

#### Desvantagens
* Pode ser computacionalmente intensivo em grandes conjuntos de dados.
* Sensível a outliers.

### Aplicações
#### Biologia Molecular:
* Agrupamento de Genes: Identificação de padrões de expressão gênica semelhantes para entender as relações entre diferentes genes.

#### Análise de Imagem e Visão Computacional:
* Segmentação de Imagens: Agrupamento de pixels semelhantes para segmentar regiões em imagens.

#### Ciências Ambientais:
* Classificação de Ecossistemas: Agrupamento de áreas geográficas semelhantes com base em características ambientais.

#### Marketing e Segmentação de Clientes:
* Segmentação de Mercado: Agrupamento de consumidores com base em padrões de comportamento de compra.

#### Neurociência e Estudos Cognitivos:
* Análise de Atividade Neural: Agrupamento de padrões de atividade neural semelhantes para entender a organização do cérebro.

#### Exemplo de código


```python
from sklearn.datasets import make_classification
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Generate synthetic data for classification
X, _ = make_classification(n_samples=300, n_features=4, n_informative=2, n_redundant=2, random_state=42)
```

```python
# Standardize the data
X = StandardScaler().fit_transform(X)
```

```python
# Apply Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
labels = agg_cluster.fit_predict(X)
```

```python
# Create a DataFrame for easy plotting
import pandas as pd
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
df['Agg Cluster Label'] = labels
```

```python
# Visualize the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage(X, method='ward'))
plt.title("Dendrogram for Agglomerative Clustering")
plt.show()
```

```python
# Plot the clustering results
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Feature 1', y='Feature 2', hue='Agg Cluster Label', data=df, palette='viridis', s=50, marker='X', edgecolor='black', legend='full')
plt.title("Agglomerative Clustering on Synthetic Classification Dataset")
plt.show()
```
