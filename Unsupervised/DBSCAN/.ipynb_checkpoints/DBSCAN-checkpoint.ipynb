# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

## O DBSCAN é um algoritmo de clustering que agrupa pontos com base na densidade espacial. Ao contrário de métodos como o K-Means, o DBSCAN não requer a especificação do número de clusters a priori.

### Como funciona:

#### Definição de Parâmetros:
* Eps (Epsilon): Raio em torno de um ponto para procurar vizinhos.
* MinPts: Número mínimo de pontos dentro de Eps para formar um cluster.

#### Atribuição de Pontos:
* Pontos dentro do Eps uns dos outros são considerados vizinhos.
* Pontos com mais vizinhos que MinPts são considerados núcleos e formam um cluster.

#### Expansão do Cluster:
* Os núcleos são expandidos adicionando pontos alcançáveis a partir de seus vizinhos.

#### Atribuição de Ruído:
* Pontos que não pertencem a nenhum cluster são considerados ruído.

![dbsacn](dbscan.png)

#### Vantagens
* Identificação de clusters de formas arbitrárias.
* Capacidade de lidar com densidades variáveis nos clusters.
* Não requer a especificação do número de clusters.

#### Desvantagens
* Sensível à escolha de parâmetros Eps e MinPts.
* Dificuldade em identificar clusters de densidades diferentes.
* Pode classificar pontos em bordas como ruído.

### Aplicações
#### Detecção de Anomalias:
* Identificação de pontos isolados ou padrões incomuns em conjuntos de dados.
* Útil em segurança cibernética para detectar atividades anômalas.

#### Monitoramento de Tráfego e Mobilidade:
* Agrupamento de dados de movimento para identificar padrões de tráfego.
* Análise de padrões de deslocamento em dados de GPS.

#### Segmentação de Imagens e Vídeos:
* Agrupamento de pixels em imagens para segmentação semântica.
* Identificação de objetos e regiões em vídeos.

#### Agrupamento de Genes em Bioinformática:
* Agrupamento de perfis de expressão gênica.
* Identificação de padrões genéticos em grandes conjuntos de dados biológicos.

#### Marketing e Comportamento do Consumidor:
* Segmentação de clientes com base em padrões de compra.
* Análise de padrões de comportamento do consumidor em dados de transações.

#### Exemplo de código com DBSCAN

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Generate synthetic data with multiple clusters
X, _ = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)

# Standardize the data
X = StandardScaler().fit_transform(X)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
# Create a DataFrame for easy plotting
# Create a DataFrame for easy plotting
import pandas as pd
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
df['DBSCAN Label'] = labels

# Plot the results
plt.figure(figsize=(10, 6))

# Plot DBSCAN cluster assignments
sns.scatterplot(x='Feature 1', y='Feature 2', hue='DBSCAN Label', data=df, palette='viridis', s=50, marker='X', edgecolor='black', legend='full')

plt.title("DBSCAN Clustering on Synthetic Dataset")
plt.show()
```
